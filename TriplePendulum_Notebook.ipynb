# Multiple-Segments Pendulum

In this Notebook we try to solve a simple computational physics problem: integrating the equations of motions of a pendulum made of several, let's say "n", segments and masses. 

In the picture below we see a "double" pendulum, a pendulum made of n=2 segments (l<sub>1</sub> & l<sub>2</sub>) and masses (m<sub>1</sub> & m<sub>2</sub>)!

## Single Pendulum

```python
n = 1

# Time grid (starting time, ending time, interval) 
t = np.linspace(0, 10, 10000) # (s)

# Initial positions (if scalar->applies to all mass points, if vector->make sure its length is n)
ip = 135 # (deg)

# Initial velocities (if scalar->applies to all mass points, if vector->make sure its length is n)
iv = 0 # (deg/s)

# Lenghts of the segments (if scalar->applies to all segments, if vector->make sure its length is n)
l = 1 # (m)

# Masses of the points (if scalar->applies to all mass points, if vector->make sure its length is n)
m = 1 # (kg)
```

```python
# Call function in module "TriplePendulum_Code" to integrate the system's equations of motion 
p = tp.integrate_pendulum(n=n, times=t, initial_positions=ip, initial_velocities=iv, lengths=l, masses=m)

# Call function in module "TriplePendulum_Code" to get (x, y) coordinates from generalized coordinates
x, y = tp.get_xy_coords(p)
```

```python
# Call function in module "TriplePendulum_Code" to show the trajectories of each mass point
fig = tp.make_plot(x, y)

#fig.savefig('./Images/SinglePendulumTrajectory.png', dpi = 300, facecolor = 'white')
```

```python
# Call function in module "TriplePendulum_Code" to make the animation of the pendulum
anim = tp.animate_pendulums(n=n, times=np.linspace(0, 10, 1000), initial_positions=ip, initial_velocities=iv, lengths=l, masses=m)

anim.save('./Videos/SinglePendulumAnimation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
```
